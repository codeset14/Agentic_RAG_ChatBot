# ðŸ§  Pandas Cheat Sheet for Data Science

```
import pandas as pd
```

## ðŸ“ 1. Load & Save Data

```
df = pd.read_csv('file.csv')           # Load CSV
df = pd.read_excel('file.xlsx')        # Load Excel
df.to_csv('output.csv', index=False)   # Save to CSV
df.to_excel('output.xlsx')             # Save to Excel
```

## ðŸ” 2. Quick Overview

```
df.head()              # First 5 rows
df.tail()              # Last 5 rows
df.shape               # (rows, columns)
df.columns             # Column names
df.info()              # Data summary
df.describe()          # Statistics
df.dtypes              # Data types
```

## ðŸ”Ž 3. Select & Filter

```
df['col']                     # Select column (Series)
df[['col1', 'col2']]          # Select multiple columns
df.iloc[0]                    # Select row by position
df.loc[0]                     # Select row by index

df[df['age'] > 30]            # Filter rows
df[(df['age'] > 30) & (df['gender'] == 'Male')]  # Multiple conditions
```

## ðŸ§¼ 4. Data Cleaning

```
df.isnull().sum()             # Count nulls
df.dropna()                   # Drop missing rows
df.fillna(0)                  # Fill missing with 0
df.fillna(df['col'].mean())   # Fill with mean
df.drop_duplicates()          # Drop duplicates
df['col'] = df['col'].astype(int)   # Convert type
```

## ðŸ§  5. Transformations

```
df['new'] = df['col'].apply(lambda x: x * 2)  # Custom logic
df['col'] = df['col'].map({'A': 1, 'B': 2})   # Replace values
df.rename(columns={'old': 'new'}, inplace=True)
df.sort_values('col', ascending=False)
```

## ðŸ”„ 6. Grouping & Aggregation

```
df.groupby('dept')['salary'].mean()
df.groupby(['dept', 'gender']).agg({
    'salary': ['mean', 'max'],
    'age': 'median'
})
```
